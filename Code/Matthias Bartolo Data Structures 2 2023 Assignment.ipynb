### 1. Creation of sets X, Y, and Z without duplicates and showing intersections:

#### Creation of Set X

```python
#Declaring set X as an unordered list, and generating p
X=list()
p=random.randint(1000, 3000)

#Looping until the length of X is smaller than p
while(len(X)<p):
    #Generating a random number, and checking whether generated number is in X or not, if not, number will be added to set
    num=random.randint(-3000, 3000)
    if(num not in X):
        X.append(num)
```

#### Showing Set Interesections:

```python
#Creating function, which returns the intersection set between the two inputted sets
def SetInteresection(firstSet, secondSet):
    #Declaring set to hold intersection
    intersectionSet=list()
    #Looping through all the elements in the first set
    for element1 in firstSet:
        #Checking whether element is in secondSet, if so appending element to intersection Set
        if(element1 in secondSet):
            intersectionSet.append(element1)
    #Returning Intersection Set
    return intersectionSet
```

### 2. Unbalanced BST 

#### Node Encoding:

**Class Variables:**
1. totalNoOfComparisons - holds the total Number of Comparisons Value
2. totalNoOfNodes - holds the total Number of Nodes for the whole Tree


**Instance Variables:**
1. value - holds the current node value
2. leftChild - holds a pointer to the current node's left Subtree
3. rightChild - holds a pointer to the current node's right Subtree

```python
class BSTNode:
    totalNoOfComparisons=0
    totalNoOfNodes=0
    
    def __init__(self, value):
        self.value = value
        self.leftChild = None
        self.rightChild = None
```

#### BST Methods:

```python
#Method which inserts a new node into the BST    
def BSTInsertion(root,value):
    #Checking whether root is None
    if(root is not None):
        #Checking whether current value is smaller than the root value
        if(value<root.value):
            #Adding to the number of total comparisons, and recursively calling BSTInsertion method on left Child
            BSTNode.totalNoOfComparisons+=1
            root.leftChild=BSTInsertion(root.leftChild,value) 
        #Checking whether current value is larger than the root value
        elif(value>root.value):
            #Adding to the number of total comparisons, and recursively calling BSTInsertion method on right Child
            BSTNode.totalNoOfComparisons+=1
            root.rightChild=BSTInsertion(root.rightChild,value)
        #Returning root
        return root
    else:
        #Creating new BST Node, and adding to the total number of nodes
        root=BSTNode(value)
        BSTNode.totalNoOfNodes+=1
        #Returning root
        return root

    
#Method which deletes a node from the BST  
def BSTDeletion(root,value):
    #Checking whether root is None, and if so returning root
    if(root is None): 
         return root
    #Checking whether current value is smaller than the root value
    if(value<root.value):
        #Adding to the number of total comparisons, and recursively calling BSTDeletion method on left Child
        BSTNode.totalNoOfComparisons+=1
        root.leftChild=BSTDeletion(root.leftChild,value)
    #Checking whether current value is larger than the root value
    elif(value>root.value):
        #Adding to the number of total comparisons, and recursively calling BSTDeletion method on right Child
        BSTNode.totalNoOfComparisons+=1
        root.rightChild=BSTDeletion(root.rightChild,value)
        #Checking whether the requested value to delete is the root value
    elif(root.value==value):
        #Decrementing total Number of Nodes in tree
        BSTNode.totalNoOfNodes-=1
        #Case 1: root is a leaf, and thus returning None to delete root
        if(root.rightChild is None and root.leftChild is None):
            return None
        #Case 2: root doesn't have a right Child, and thus returning leftChild to delete root
        elif(root.rightChild is None):
            return root.leftChild
        #Case 3: root doesn't have a left Child, and thus returning rightChild to delete root
        elif(root.leftChild is None):
            return root.rightChild
        #Case 4: root has both left and right Children
        else:
            #Using Helper function, to get the smallest child in the right Subtree
            tmp=GetMin(root.rightChild)
            #Setting the root.value to the smallest child value
            root.value=tmp.value
            #Recursively calling BSTDeletion method on right Child, to delete smallest child, from the right Subtree
            root.rightChild=BSTDeletion(root.rightChild,tmp.value)
            #Incrementing the total Number Of Nodes to factor for the deletion of the smallest child
            BSTNode.totalNoOfNodes+=1
    #Returning root
    return root

#Method which searches for a value in the Tree
def BSTSearch(root,value):
    #If root is None, then returning false as value is not present in Tree
    if(root is None ):
        return False
    #If root value is requested value, then returning true as value is present in Tree
    elif(root.value == value):
        return True
    #If root is a leaf node, then returning false as value is not present in Tree
    elif(root.leftChild is None and root.rightChild is None):
        return False
    #Checking whether current value is smaller than the root value
    elif(value<root.value):
        #Adding to the number of total comparisons, and recursively calling BSTSearch method on left Child
        BSTNode.totalNoOfComparisons+=1
        return BSTSearch(root.leftChild,value)
    #Checking whether current value is larger than the root value
    elif(value>root.value):
        #Adding to the number of total comparisons, and recursively calling BSTSearch method on right Child
        BSTNode.totalNoOfComparisons+=1
        return BSTSearch(root.rightChild,value)

#Method which returns the smallest value in the Tree
def GetMin(root):
    #Checking whether left Child is None, if so will return root, if not recursively calling function on left Child
    if(root.leftChild is None):
        return root
    else:
        return GetMin(root.leftChild)

#Method which calculates the Tree Height
def GetTreeHeight(root):
    #If root is None, or root is a leaf node, then return 0
    if(root is None or root.rightChild is None and root.leftChild is None):
        return 0
    #Recursively calling GetTreeHeight function on the left and right subtree, and returning 1 + the max of the results
    return 1+max(GetTreeHeight(root.leftChild),GetTreeHeight(root.rightChild))
```

### 3. AVL Tree

#### Node Encoding:

**Class Variables:**
1. totalNoOfComparisons - holds the total Number of Comparisons Value
2. totalNoOfNodes - holds the total Number of Nodes for the whole Tree
3. totalNoOfRotations - holds the total Number of Rotations for the whole Tree


**Instance Variables:**
1. value - holds the current node value
2. leftChild - holds a pointer to the current node's left Subtree
3. rightChild - holds a pointer to the current node's right Subtree
4. height -  holds the height of the current node

```python
class AVLNode:
    totalNoOfComparisons=0
    totalNoOfNodes=0
    totalNoOfRotations=0
    
    def __init__(self,value):
        self.value = value
        self.leftChild = None
        self.rightChild = None
        self.height= 0
```

#### AVL Tree Methods:

```python
#Creating AVL Tree Data Structure
class AVLTree():
    def __init__(self):
        self.TNULL = AVLNode(0)
        self.treeRoot = self.TNULL
        self.TNULL.leftChild = None
        self.TNULL.rightChild = None
        
    #Method which inserts a new node into the BST    
    def AVLInsertion(self,value):
        #Setting parent to None and iterator to treeRoot
        parent=None
        iterator = self.treeRoot

        #Iterating through the tree, until a null node is found
        while (iterator != self.TNULL):
            #Setting the parent to be the iteratos
            parent = iterator
            #Checking whether current value is smaller than the iterator value
            if (value < iterator.value):
                #Incrementing the total Number of Comparisons, and setting iterator to leftChild
                AVLNode.totalNoOfComparisons+=1
                iterator = iterator.leftChild
            #Checking whether current value is larger than the iterator value
            else:
                #Incrementing the total Number of Comparisons, and setting iterator to rightChild
                AVLNode.totalNoOfComparisons+=1
                iterator = iterator.rightChild

        #Creating new AVL Node, and setting the respective variables
        newNode = AVLNode(value)
        newNode.leftChild = self.TNULL
        newNode.rightChild = self.TNULL
        #Incrementing the total number of nodes
        BSTNode.totalNoOfNodes+=1
        #Checking whether parent is None, i.e., insertion of root
        if (parent == None):
            self.treeRoot = newNode
            return
        #Checking whether current value is smaller than the parent value and setting the respective pointer to point to new node
        elif (value < parent.value):
            parent.leftChild = newNode
        #Checking whether current value is larger than the parent value and setting the respective pointer to point to new node
        else:
            parent.rightChild = newNode
    
    #Method which searches for a value in the Tree
    def BSTSearch(self,node,value):
        #If node is None, then returning false as value is not present in Tree
        if(node is self.TNULL ):
            return False
        #If node value is requested value, then returning true as value is present in Tree
        elif(node.value == value):
            return True
        #If node is a leaf node, then returning false as value is not present in Tree
        elif(node.leftChild is self.TNULL and node.rightChild is self.TNULL):
            return False
        #Checking whether current value is smaller than the node value
        elif(value<node.value):
            #Adding to the number of total comparisons, and recursively calling BSTSearch method on left Child
            BSTNode.totalNoOfComparisons+=1
            return self.BSTSearch(node.leftChild,value)
        #Checking whether current value is larger than the node value
        elif(value>node.value):
            #Adding to the number of total comparisons, and recursively calling BSTSearch method on right Child
            BSTNode.totalNoOfComparisons+=1
            return self.BSTSearch(node.rightChild,value)
        
            
    #Method which returns the node with the smallest value in the tree
    def GetMin(self, node):
        #Looping until node.leftChild is not None, and setting node to node.leftChild
        while node.leftChild != self.TNULL:
            node = node.leftChild
        #Returning node with the smallest value
        return node
    
    #Method which calculates the Tree Height
    def GetTreeHeight(self,root):
        #If root is None, or root is a leaf node, then return 0
        if(root is self.TNULL or root.rightChild is self.TNULL and root.leftChild is self.TNULL):
            return 0
        #Recursively calling GetTreeHeight function on the left and right subtree, and returning 1 + the max of the results
        return 1+max(self.GetTreeHeight(root.leftChild),self.GetTreeHeight(root.rightChild))
```

```python
#Method which inserts a new node into the AVL Tree    
def AVLInsertion(root,value):
    #Checking whether root is None
    if(root is not None):
        #Checking whether current value is smaller than the root value
        if(value<root.value):
            #Adding to the number of total comparisons, and recursively calling AVLInsertion method on left Child
            AVLNode.totalNoOfComparisons+=1
            root.leftChild=AVLInsertion(root.leftChild,value) 
        #Checking whether current value is larger than the root value
        elif(value>root.value):
            #Adding to the number of total comparisons, and recursively calling AVLInsertion method on right Child
            AVLNode.totalNoOfComparisons+=1
            root.rightChild=AVLInsertion(root.rightChild,value)
        
        #Calling the CheckReBalancing method
        root=CheckReBalancing(root)
        #Returning root
        return root
    else:
        #Creating new AVL Node, and adding to the total number of nodes
        root=AVLNode(value)
        AVLNode.totalNoOfNodes+=1
        #Returning root
        return root

#Method which deletes a node from the AVL Tree         
def AVLDeletion(root,value):
    #Checking whether root is None, and if so returning root
    if(root is None): 
         return root
    #Checking whether current value is smaller than the root value
    if(value<root.value):
        #Adding to the number of total comparisons, and recursively calling AVLDeletion method on left Child
        AVLNode.totalNoOfComparisons+=1
        root.leftChild=AVLDeletion(root.leftChild,value)
    #Checking whether current value is larger than the root value
    elif(value>root.value):
        #Adding to the number of total comparisons, and recursively calling AVLDeletion method on right Child
        AVLNode.totalNoOfComparisons+=1
        root.rightChild=AVLDeletion(root.rightChild,value)
    #Checking whether the requested value to delete is the root value
    elif(root.value==value):
        #Decrementing total Number of Nodes in tree
        AVLNode.totalNoOfNodes-=1
        #Case 1: root is a leaf, and thus returning None to delete root
        if(root.rightChild is None and root.leftChild is None):
            return None
        #Case 2: root doesn't have a right Child, and thus returning leftChild to delete root
        elif(root.rightChild is None):
            return root.leftChild
        #Case 3: root doesn't have a left Child, and thus returning rightChild to delete root
        elif(root.leftChild is None):
            return root.rightChild
        #Case 4: root has both left and right Children
        else:
            #Using Helper function, to get the smallest child in the right Subtree
            tmp=GetMin(root.rightChild)
            #Setting the root.value to the smallest child value and reducing height
            root.value=tmp.value
            root.height-=1
            #Recursively calling AVLDeletion method on right Child, to delete smallest child, from the right Subtree
            root.rightChild=AVLDeletion(root.rightChild,tmp.value)
            #Incrementing the total Number Of Nodes to factor for the deletion of the smallest child
            AVLNode.totalNoOfNodes+=1
            
    #Calling the CheckReBalancing method
    root=CheckReBalancing(root)
    #Returning root
    return root
    
#Method which searches for a value in the Tree
def AVLSearch(root,value):
    #If root is None, then returning false as value is not present in Tree
    if(root is None ):
        return False
    #If root value is requested value, then returning true as value is present in Tree
    elif(root.value == value):
        return True
    #If root is a leaf node, then returning false as value is not present in Tree
    elif(root.leftChild is None and root.rightChild is None):
        return False
    #Checking whether current value is smaller than the root value
    elif(value<root.value):
        #Adding to the number of total comparisons, and recursively calling AVLSearch method on left Child
        AVLNode.totalNoOfComparisons+=1
        return AVLSearch(root.leftChild,value)
    #Checking whether current value is larger than the root value
    elif(value>root.value):
        #Adding to the number of total comparisons, and recursively calling AVLSearch method on right Child
        AVLNode.totalNoOfComparisons+=1
        return AVLSearch(root.rightChild,value)

#Method which returns the AVL Height for a specific root
def GetAVLHeight(root):
    #Checking whether root is None, if so returning -1
    if(root is None):
        return -1
    #Checking whether root is a leaf node, if so returning 0
    elif(root.leftChild is None and root.rightChild is None):
        return 0
    #Checking whether root doesn't have a left Child, if so returning 1 + height of right Subtree
    elif(root.leftChild is None):
        return 1 +root.rightChild.height
    #Checking whether root doesn't have a right Child, if so returning 1 + height of left Subtree
    elif(root.rightChild is None):
        return 1 +root.leftChild.height
    #Else returning 1 + max height of left Subtree and right Subtree
    else:
        return 1 + max(root.leftChild.height,root.rightChild.height)

#Method which returns the balance factor for a specific root
def GetBalanceFactor(root):
    #Checking whether root is None or root is leaf node, if so returning 0
    if(root is None or (root.leftChild is None and root.rightChild is None)):
        return 0
    #Else returning height of left Subtree - height of right Subtree
    else:
        return GetAVLHeight(root.leftChild)-GetAVLHeight(root.rightChild)

#Method which checks whether tree needs rebalancing
def CheckReBalancing(root):
    #Retrieving the root height through GetAVLHeight method
    root.height=GetAVLHeight(root)
    #Calculating balance Factor for root
    balanceFactor = GetBalanceFactor(root)
    #Checking whether balanceFactor is less than -1
    if(balanceFactor < -1):
        #Checking whether balanceFactor of rightChild is smaller than or equal to 0
        if(GetBalanceFactor(root.rightChild) <= 0):
            #Applying a Right Right rotation
            root = RightRotationAVL(root)
        #Checking whether balanceFactor of rightChild is larger than 0
        elif(GetBalanceFactor(root.rightChild) > 0):
            #Applying a Right Left rotation
            root.rightChild = LeftRotationAVL(root.rightChild)
            root = RightRotationAVL(root)
        #Incrementing total No of Rotations
        AVLNode.totalNoOfRotations+=1
    #Else checking whether balanceFactor is less than -1
    elif(balanceFactor > 1):
        #Checking whether balanceFactor of leftChild is larger than or equal to 0
        if(GetBalanceFactor(root.leftChild) >= 0):
            #Applying a Left Left rotation
            root = LeftRotationAVL(root)
        #Checking whether balanceFactor of rightChild is smaller than 0
        elif(GetBalanceFactor(root.leftChild) < 0):
            # Applying a Left Right rotation
            root.leftChild = RightRotationAVL(root.leftChild)
            root = LeftRotationAVL(root)
        #Incrementing total No of Rotations
        AVLNode.totalNoOfRotations+=1
    #Returning root
    return root

#Method which perform an AVL Left Rotation on a root            
def LeftRotationAVL(root):
    #Checking if root is None, if so returning
    if(root is None):
        return
    #Checking whether leftChild is None, if so returning root
    if(root.leftChild is None):
        return root
    #Setting y to hold leftChild of root
    y=root.leftChild
    #Setting B to hold rightChild of y
    B=y.rightChild
    #Setting leftChild of root to B
    root.leftChild=B
    #Setting rightChild of y to root
    y.rightChild=root
    #Calculating new node heights
    root.height=GetAVLHeight(root)
    y.height=GetAVLHeight(y)
    #Returning y
    return y

#Method which perform an AVL Right Rotation on a root 
def RightRotationAVL(root):
    #Checking if root is None, if so returning
    if(root is None):
        return
    #Checking whether rightChild is None, if so returning root
    if(root.rightChild is None):
        return root
    #Setting y to hold rightChild of root
    y=root.rightChild
    #Setting B to hold leftChild of y
    B=y.leftChild
    #Setting rightChild of root to B
    root.rightChild=B
    #Setting leftChild of y to root
    y.leftChild=root
    #Calculating new node heights
    root.height=GetAVLHeight(root)
    y.height=GetAVLHeight(y)
    #Returning y
    return y
```

### 4. Red Black Tree

#### Node Encoding:

**Class Variables:**
1. totalNoOfComparisons - holds the total Number of Comparisons Value
2. totalNoOfNodes - holds the total Number of Nodes for the whole Tree
3. totalNoOfRotations - holds the total Number of Rotations for the whole Tree


**Instance Variables:**
1. value - holds the current node value
2. leftChild - holds a pointer to the current node's left Subtree
3. rightChild - holds a pointer to the current node's right Subtree
4. parent -  holds a pointer to the current node's parent
5. color - holds the color bit for the current Node ("1" marks for black color, and "0" marks for red color)

**Important Notice:**
- The implementation of the Red Black Tree varies slightly from the BST and AVL implementations, as creation of such a tree required the use of the TNULL, in order to make the code more robust, whilst avoid the hardcoding of numerous checks, for instance whether node is None.

```python
class RedBlackTreeNode:
    totalNoOfComparisons=0
    totalNoOfNodes=0
    totalNoOfRotations=0
    
    def __init__(self,value):
        self.value = value
        self.leftChild = None
        self.rightChild = None
        self.parent=None
        self.color=0
```

#### Red Black Tree Methods:

```python
#Creating Red Black Tree Data Structure
class RedBlackTree():
    def __init__(self):
        self.TNULL = RedBlackTreeNode(0)
        self.treeRoot = self.TNULL
        self.TNULL.leftChild = None
        self.TNULL.rightChild = None
        self.TNULL.color = 1
               
    #Method which inserts a new node into the Red Black Tree  
    def RBTInsertion(self, value):
        #Initialising parent to None and iterator to root of tree
        parent = None
        iterator = self.treeRoot
        
        #Iterating through the tree, until a null node is found
        while (iterator != self.TNULL):
            #Setting the parent to be the iteratos
            parent = iterator
            #Checking whether current value is smaller than the iterator value
            if (value < iterator.value):
                #Incrementing the total Number of Comparisons, and setting iterator to leftChild
                RedBlackTreeNode.totalNoOfComparisons+=1
                iterator = iterator.leftChild
            #Checking whether current value is larger than the iterator value
            else:
                #Incrementing the total Number of Comparisons, and setting iterator to rightChild
                RedBlackTreeNode.totalNoOfComparisons+=1
                iterator = iterator.rightChild
        
        #Creating new Red Black Tree Node, and setting the respective variables and counter
        newNode = RedBlackTreeNode(value)
        newNode.parent = parent
        newNode.leftChild = self.TNULL
        newNode.rightChild = self.TNULL
        #Incrementing the total number of nodes
        RedBlackTreeNode.totalNoOfNodes+=1
        #Checking whether parent is None, i.e., insertion of root, thus color needs to be black
        if (parent == None):
            self.treeRoot = newNode
            newNode.color = 1
            return
        #Checking whether current value is smaller than the parent value and setting the respective pointer to point to new node
        elif (value < parent.value):
            parent.leftChild = newNode
        #Checking whether current value is larger than the parent value and setting the respective pointer to point to new node
        else:
            parent.rightChild = newNode
            
        #If grandparent of new node is None, which means child of treeRoot, returning, as there is no need to repair the tree
        if newNode.parent.parent == None:
            return
        #Calling the RepairTreeInsert on newNode
        self.RepairTreeInsert(newNode)
    
    #Method which deletes a value from the Tree
    def RBTDeletion(self,value):
        #Initialising nodeToDelete to None and iterator to treeRoot
        nodeToDelete = self.TNULL
        iterator = self.treeRoot

        #Iterating through the tree, until a null node is found
        while iterator != self.TNULL:
            #Checking whether value is the iterator's value
            if(value==iterator.value):
                #Setting the parent to be the iteratos
                nodeToDelete = iterator
                break
            #Checking whether current value is smaller than the iterator value
            if (value < iterator.value):
                #Incrementing the total Number of Comparisons, and setting iterator to leftChild
                RedBlackTreeNode.totalNoOfComparisons+=1
                iterator = iterator.leftChild
            #Checking whether current value is larger than the iterator value
            else:
                #Incrementing the total Number of Comparisons, and setting iterator to rightChild
                RedBlackTreeNode.totalNoOfComparisons+=1
                iterator = iterator.rightChild
        
        #Checking whether nodeToDelete is None, if so returning as value is not found
        if(nodeToDelete == self.TNULL):
            return
        #Else value is found, and must delete the value
        else:
            #Decrementing the total Number of Nodes
            RedBlackTreeNode.totalNoOfNodes-=1
            #Storing the nodeToDelete and nodeToDeleteColor in 2 variables
            nodeToDeleteCopy=nodeToDelete
            nodeToDeleteColor=nodeToDeleteCopy.color
            #Checking whether nodeToDelete doesn't have a leftChild
            if(nodeToDelete.leftChild is self.TNULL):
                #Creating new variable tmp to hold to rightChild of nodeToDelete and Transplanting rightChild in the position of
                #nodeToDelete
                tmp=nodeToDelete.rightChild
                self.Transplant(nodeToDelete,nodeToDelete.rightChild)
            #Checking whether nodeToDelete doesn't have a rightChild
            elif(nodeToDelete.rightChild is self.TNULL):
                #Creating new variable tmp to hold to leftChild of nodeToDelete and Transplanting leftChild in the position of
                #nodeToDelete
                tmp=nodeToDelete.leftChild
                self.Transplant(nodeToDelete,nodeToDelete.leftChild)
            #Else nodeToDelete has bothe left and right subtrees
            else:
                #Retrieving the smallest value from the right subtree, and storing node in nodeToDeleteCopy
                nodeToDeleteCopy=self.GetMin(nodeToDelete.rightChild)
                nodeToDeleteColor=nodeToDeleteCopy.color
                #Setting tmp to nodeToDeleteCopy.rightChild
                tmp=nodeToDeleteCopy.rightChild
                #Checking whether parent of nodeToDeleteCopy is nodeToDelete
                if(nodeToDeleteCopy.parent is nodeToDelete):
                    #Setting tmp parent to nodeToDeleteCopy
                    tmp.parent=nodeToDeleteCopy
                else:
                    #Else Transplanting rightChild in the position of nodeToDeleteCopy
                    self.Transplant(nodeToDeleteCopy,nodeToDeleteCopy.rightChild)
                    #Changing the relevant pointers of the rightChild of nodeToDeleteCopy
                    nodeToDeleteCopy.rightChild=nodeToDelete.rightChild
                    nodeToDeleteCopy.rightChild.parent=nodeToDeleteCopy
                
                #Transplanting nodeToDeleteCopy in the position of nodeToDelete
                self.Transplant(nodeToDelete,nodeToDeleteCopy)
                #Changing the relevant pointers and color of the leftChild of nodeToDeleteCopy
                nodeToDeleteCopy.leftChild=nodeToDelete.leftChild
                nodeToDeleteCopy.leftChild.parent=nodeToDeleteCopy
                nodeToDeleteCopy.color=nodeToDelete.color
            #Checking if node To Delete was black, if so calling RepairTreeDelete Method
            if(nodeToDeleteColor == 1):
                self.RepairTreeDelete(tmp)
                
    #Method which searches for a value in the Tree
    def RBTSearch(self,node,value):
        #If node is None, then returning false as value is not present in Tree
        if(node is self.TNULL ):
            return False
        #If node value is requested value, then returning true as value is present in Tree
        elif(node.value == value):
            return True
        #If node is a leaf node, then returning false as value is not present in Tree
        elif(node.leftChild is self.TNULL and node.rightChild is self.TNULL):
            return False
        #Checking whether current value is smaller than the node value
        elif(value<node.value):
            #Adding to the number of total comparisons, and recursively calling RBTSearch method on left Child
            RedBlackTreeNode.totalNoOfComparisons+=1
            return self.RBTSearch(node.leftChild,value)
        #Checking whether current value is larger than the node value
        elif(value>node.value):
            #Adding to the number of total comparisons, and recursively calling RBTSearch method on right Child
            RedBlackTreeNode.totalNoOfComparisons+=1
            return self.RBTSearch(node.rightChild,value)
        
    #Method which returns the Uncle of a node               
    def GetUncle(self,node):
        #Calling the GetSibling method on parent
        return self.GetSibling(node.parent)

    #Method which returns the Sibling of a node
    def GetSibling(self,node):
        #Checking whether node is None or parent is None, if so returning node
        if(node is None or node.parent is None):
            return node
        #Returning parent.rightChild if node is parent.leftChild
        if(node==node.parent.leftChild):
            return node.parent.rightChild
        #Returning parent.leftChild if node is parent.rightChild
        if(node==node.parent.rightChild):
            return node.parent.leftChild
        
    #Method which perform an RBT Left Rotation on a node       
    def LeftRotationRBT(self,node):
        #Setting y to hold leftChild of root
        y=node.leftChild
        #Setting node.leftChild to y.rightChild
        node.leftChild=y.rightChild
        #Checking whether leftChild is None, if so setting node.leftChild.parent to node
        if(y.rightChild != self.TNULL):
            y.rightChild.parent=node
        #Setting y.parent to node.parent
        y.parent=node.parent
        #Setting the respective node parents based on relevant statements
        if(node.parent==None):
            self.treeRoot=y
        elif(node==node.parent.rightChild):
            node.parent.rightChild=y
        else:
            node.parent.leftChild=y
        #Setting rightChild of y to root and y.rightChild.parent to y
        y.rightChild=node
        node.parent=y

    #Method which perform an RBT Right Rotation on a root 
    def RightRotationRBT(self,node):
        #Setting y to hold leftChild of root
        y=node.rightChild
        #Setting node.leftChild to y.rightChild
        node.rightChild=y.leftChild
        #Checking whether leftChild is None, if so setting node.leftChild.parent to node
        if(y.leftChild != self.TNULL):
            y.leftChild.parent=node
        #Setting y.parent to node.parent
        y.parent=node.parent
        #Setting the respective node parents based on relevant statements
        if(node.parent==None):
            self.treeRoot=y
        elif(node==node.parent.rightChild):
            node.parent.rightChild=y
        else:
            node.parent.leftChild=y
        #Setting rightChild of y to root and y.rightChild.parent to y
        y.leftChild=node
        node.parent=y
    
    #Method which transplants a newChild in the tree in the position of the oldChild
    def Transplant(self, oldChild, newChild):
        #Setting the newChild as the treeRoot if the oldChild is the root
        if (oldChild.parent == None):
            self.treeRoot = newChild
        #Checking whether oldChild is the leftChild of its parent, if so setting the relevant pointer to point to the newChild
        elif (oldChild == oldChild.parent.leftChild):
            oldChild.parent.leftChild = newChild
        #Checking whether oldChild is the rightChild of its parent, if so setting the relevant pointer to point to the newChild
        else:
            oldChild.parent.rightChild = newChild
        #Setting the newChild parent
        newChild.parent = oldChild.parent
    
    #Method which repairs tree after insert
    def RepairTreeInsert(self,node):  
        #Setting parent and grandparent to None
        parent=None
        grandparent=None

        #Looping until parent is not None and parent.color is red
        while(node.parent is not None and node.parent.color==0):
            #Retrieving node uncle
            uncle=self.GetUncle(node)
            #Initialising the parent and grandparent variables
            parent =node.parent
            grandparent=parent.parent

            #Case 1: Recoloring when uncle is red
            #Checking whether uncle and parent have a red color, and grandparent has a black color
            if(uncle.color==0 and parent.color==0 and grandparent.color==1):
                #Setting uncle and parent colors to black, and grandparent color to red
                uncle.color=1
                parent.color=1
                grandparent.color=0
                #Setting node to grandparent
                node=grandparent
            #Checking whether root color is red, and parent color is red (Red Red Violation)   
            elif(node.color==0 and parent.color==0):
                #Checking whether parent is the rightChild of the grandparent
                if(parent==grandparent.rightChild):
                    #Incrementing the total number of rotations
                    RedBlackTreeNode.totalNoOfRotations+=1
                    #Case 2: Double Rotation (checking whether node is the leftChild of the parent)
                    if(node==parent.leftChild):
                        #Setting node to node.parent
                        node=node.parent
                        #Performing Left Rotation on node (parent)
                        self.LeftRotationRBT(node)

                    #Changing grandparent and parent colors
                    node.parent.color=1
                    node.parent.parent.color=0
                    #Case 3: Single Rotation, performing Right Rotation on grandparent   
                    self.RightRotationRBT(node.parent.parent)

                #Checking whether parent is the leftChild of the grandparent
                else:
                    #Incrementing the total number of rotations
                    RedBlackTreeNode.totalNoOfRotations+=1
                    #Case 2: Double Rotation (checking whether node is the rightChild of the parent)
                    if(node==parent.rightChild):
                        #Setting node to node.parent
                        node=node.parent
                        #Performing Right Rotation on node (parent)
                        self.RightRotationRBT(node)

                    #Changing grandparent and parent colors
                    node.parent.color=1
                    node.parent.parent.color=0
                    #Case 3: Single Rotation, performing Left Rotation on grandparent 
                    self.LeftRotationRBT(node.parent.parent)

            else:
                #Setting node to grandparent
                node=grandparent

        #Case 0: Setting treeRoot to black 
        self.treeRoot.color=1
    
    #Method which returns the node with the smallest value in the tree
    def GetMin(self, node):
        #Looping until node.leftChild is not None, and setting node to node.leftChild
        while node.leftChild != self.TNULL:
            node = node.leftChild
        #Returning node with the smallest value
        return node
    
    #Method which repairs tree after delete
    def RepairTreeDelete(self, node):
        #Looping until node is not treeRoot and node doesn't have a black color
        while node != self.treeRoot and node.color == 1:
            #Checking if node is the leftChild of it's parent
            if node == node.parent.leftChild:
                #Retrieving the node's sibling
                sibling = node.parent.rightChild
                #Case 1: Sibling is red, checking whether sibling is red
                if sibling.color == 0:
                    #Setting sibling color to black and parent to red color
                    sibling.color = 1
                    node.parent.color = 0
                    #Applying RightRotation on parent and incrementing the total number of rotations
                    self.RightRotationRBT(node.parent)
                    RedBlackTreeNode.totalNoOfRotations+=1
                    #Setting sibling to be the node.parent.rightChild
                    sibling = node.parent.rightChild
                #Case 2: Both children of Sibling are black, checking whether sibling children are black
                if sibling.leftChild.color == 1 and sibling.rightChild.color == 1:
                    #Setting sibling color to red, and setting node to node.parent
                    sibling.color = 0
                    node=node.parent
                else:
                    #Incrementing the total number of rotations
                    RedBlackTreeNode.totalNoOfRotations+=1
                    #Case 3: Sibling rightChild is black, checking whether sibling rightChild is black
                    if sibling.rightChild.color == 1:
                        #Setting sibling leftChild to black and sibling to red
                        sibling.leftChild.color = 1
                        sibling.color = 0
                        #Applying LeftRotation on sibling and setting sibling to be the node.parent.rightChild
                        self.LeftRotationRBT(sibling)
                        sibling = node.parent.rightChild
                    #Case 4: Else do the following:
                    #Set the sibling to the color of its parent, parent color to black, and sibling rightChild to black
                    sibling.color = node.parent.color
                    node.parent.color = 1
                    sibling.rightChild.color = 1
                    #Applying RightRotation on parent and setting node to treeRoot
                    self.RightRotationRBT(node.parent)
                    node = self.treeRoot

            #Else node is the rightChild of it's parent
            else:
                #Retrieving the node's sibling
                sibling = node.parent.leftChild
                #Case 1: Sibling is red, checking whether sibling is red
                if sibling.color == 0:
                    #Setting sibling color to black and parent to red color
                    sibling.color = 1
                    node.parent.color = 0
                    #Applying LeftRotation on parent and incrementing the total number of rotations
                    self.LeftRotationRBT(node.parent)
                    RedBlackTreeNode.totalNoOfRotations+=1
                    #Setting sibling to be the node.parent.leftChild
                    sibling = node.parent.leftChild
                #Case 2: Both children of Sibling are black, checking whether sibling children are black
                if sibling.leftChild.color == 1 and sibling.rightChild.color == 1:
                    #Setting sibling color to red, and setting node to node.parent
                    sibling.color = 0
                    node=node.parent
                else:
                    #Incrementing the total number of rotations
                    RedBlackTreeNode.totalNoOfRotations+=1
                    #Case 3: Sibling leftChild is black, checking whether sibling leftChild is black
                    if sibling.leftChild.color == 1:
                        #Setting sibling rightChild to black and sibling to red
                        sibling.rightChild.color = 1
                        sibling.color = 0
                        #Applying RightRotation on sibling and setting sibling to be the node.parent.leftChild
                        self.RightRotationRBT(sibling)
                        sibling = node.parent.leftChild
                    #Case 4: Else do the following:
                    #Set the sibling to the color of its parent, parent color to black, and sibling leftChild to black
                    sibling.color = node.parent.color
                    node.parent.color = 1
                    sibling.leftChild.color = 1
                    #Applying LeftRotation on parent and setting node to treeRoot
                    self.LeftRotationRBT(node.parent)
                    node = self.treeRoot
        #Setting node color to black
        node.color = 1
```
